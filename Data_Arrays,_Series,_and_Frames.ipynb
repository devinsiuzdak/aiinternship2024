# Creating an Array

```python
#float64 doubles the precisision of floats to make them more precise.
floatdata= [5.0,9.0,6.0,8.0]
floatdata1  = [1.0,2.0,3.0,4.0]
floatarr2= np.float64([floatdata, floatdata1])
floatarr2
```

```python
floatarr2.shape
```

```python

# Create a NumPy array of float64 data type
arr = np.array([1.23456789, 9.87654321], dtype=np.float64)

# Print the data type of the array
print(arr.dtype)
print(arr)
```

```python
#Creating an empty np
empty_arr=np.empty((2,3))
empty_arr
```

```python
#Creating Range
range_arr=np.arange(data1[3])
range_arr
```

```python
#Interval: [2, 9)
#Step size: 2

#2 -> 4 -> 6 -> 8 -> 9 (exclusive)
interval_arr=np.arange(2,9,2)
interval_arr
```

```python
#Interval: [2, 9]
#Number of elements: 5

#2 -> 3.5 -> 5 -> 6.5 -> 8 -> 9
lined_arr=np.linspace(2,9,5)
lined_arr
```
