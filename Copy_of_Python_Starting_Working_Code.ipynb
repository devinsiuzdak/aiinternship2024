# Section 2 - Practice

```python
# Put all pixels in black
# RGB = 0,0,0

for x in range(my_image.width):
  for y in range(my_image.height):
    my_image.putpixel((x,y), (0,0,0))

my_image
```

```python
# Put all pixels where x and y are even in black
# RGB = 0,0,0

for x in range(my_image.width):
  for y in range(my_image.height):
    if x % 2 == 0 and y % 2 == 0:
      my_image.putpixel((x,y), (0,0,0))

my_image

```

```python
# Pixel average
# TOCOMPLETE
def pixel_average(p):
  return (p[0]+p[1]+p[2])/3
```

```python
# Replace bright pixels by grey pixels
# TOCOMPLETE
for x in range(my_image.width):
  for y in range(my_image.height):
    p = my_image.getpixel((x,y))
    if pixel_average(p) > 158:
      my_image.putpixel((x,y), (120,120,120))


my_image
```

```python
for x in range (my_image.width):
  for y in range (my_image.height):
    p = my_image.getpixel((x,y))
    if pixel_average(p) > 158:
      my_image.putpixel((x,y), (252,47,139))

my_image
```

```python
new_img = Image.new('RGB', (my_image.width+20, my_image.height+20))
new_img

for x in range(my_image.width):
  for y in range(my_image.height):
    new_img.putpixel((x+10,y+10), my_image.getpixel((x,y)))

new_img
```
